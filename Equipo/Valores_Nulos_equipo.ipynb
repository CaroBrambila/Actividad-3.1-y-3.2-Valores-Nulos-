# **Datos de Facturación**



```python
df = pd.read_excel('Datos de Facturación (1).xlsx')
df.head()
```

```python
##Identificar los valores nulos
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
data=df.copy()
```

```python
#Rellenar valores nulos con el VALOR ANTERIOR (forward fill)
##Pq va de uno en uno pero como son faltantes
df["CVE_DOC"] = df["CVE_DOC"].fillna(method="ffill")
#Como son fechas debe y estan registradas por fecha y estan en orden
df["FECHAELAB"] = df["FECHAELAB"].fillna(method="ffill")
df["FECHA_DOC"] = df["FECHA_DOC"].fillna(method="ffill")
```

```python
#Sustituir valores nulos con un VALOR ESPECÍFICO (or ejemplo, 0)
##Eligí por "C" ya que la mayoría es E
df["STATUS"] = df["STATUS"].fillna("C")
df["FECHA_CANCELA"] = df["FECHA_CANCELA"].fillna("09/09/2019")
```

```python
#Sustituir valores nulos con la MEDIA
#Promedio (mean): Se utiliza para un conjunto de datos más equilibrado sin outliers.
df["CVE_VEND"] = df["CVE_VEND"].fillna(df["CVE_VEND"].mean())
df["CAN_TOT"] = df["CAN_TOT"].fillna(df["CAN_TOT"].mean())
```

```python
#Sustituir valores nulos con la MEDIANA
##Mediana (median): Se utiliza cuando los datos tienen outliers o una distribución sesgada, ya que es menos sensible a valores extremos.
df["DES_TOT"] = df["DES_TOT"].fillna(df["DES_TOT"].median())
df["DES_FIN"] = df["DES_FIN"].fillna(df["DES_FIN"].median())
```

```python
##string en concreto
df["FECHA_ENT"]=df["FECHA_ENT"].fillna("09/09/2019  12:00:00 a. m.")
```

```python
##Identificar los valores nulos
valores_nulos=df.isnull().sum()
valores_nulos
```

# **Detalle precios y productos fabricados 2022**

```python
productos=pd.read_excel('Detalle precios y productos fabricados 2022 (2).xlsx')
productos.head()
```

```python
##Identificar los valores nulos
valores_nulos=productos.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos con la MEDIA
#Promedio (mean): Se utiliza para un conjunto de datos más equilibrado sin outliers.
productos["CANT"] = productos["CANT"].fillna(productos["CANT"].mean())
productos["COSTO_UNITARIO"] = productos["COSTO_UNITARIO"].fillna(productos["COSTO_UNITARIO"].mean())
productos["SUBTOTAL_PARTIDA"] = productos["SUBTOTAL_PARTIDA"].fillna(productos["SUBTOTAL_PARTIDA"].mean())
```

```python
#Sustituir valores nulos con la MEDIANA
##Mediana (median): Se utiliza cuando los datos tienen outliers o una distribución sesgada, ya que es menos sensible a valores extremos.
productos["PRECIO_UNITARIO"] = productos["PRECIO_UNITARIO"].fillna(productos["PRECIO_UNITARIO"].median())
productos["COSTO_UNITARIO_CALCULADO"] = productos["COSTO_UNITARIO_CALCULADO"].fillna(productos["COSTO_UNITARIO_CALCULADO"].median())
```

```python
#Rellenar valores nulos con el VALOR ANTERIOR (forward fill)
productos["CVE_DOC"] = productos["CVE_DOC"].fillna(method="ffill")
productos["NOMBRE_VENDEDOR"] = productos["NOMBRE_VENDEDOR"].fillna(method="ffill")
productos["NOMBRE_CLIENTE"] = productos["NOMBRE_CLIENTE"].fillna(method="ffill")
productos["FECHA_DOC"] = productos["FECHA_DOC"].fillna(method="ffill")
```

```python
#DESCR / BAN MICA 120mmØx50mm 850W 440V TTORNILLO
##string en concreto
productos["DESCR"]=productos["DESCR"].fillna("BAN MICA 120mmØx50mm 850W 440V TTORNILLO")
productos["CVE_ART"]=productos["CVE_ART"].fillna("TUA00406")
```

```python
##Identificar los valores nulos
valores_nulos=productos.isnull().sum()
valores_nulos
```
