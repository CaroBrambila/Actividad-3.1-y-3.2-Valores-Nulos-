```python
##Carga desde un archivo .csv sin indice
data=pd.read_csv("Ventas_totales.csv")
data.head(3)
```

```python
##Identificar los valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

METODOS DE SUSTITUCIÓN

```python
##creamos una copia del dataframe
data2=data1.copy()
data2
```

```python
#Sustituir valores nulos con la MEDIA
#Promedio (mean): Se utiliza para un conjunto de datos más equilibrado sin outliers.
data["salon_ventas"] = data["salon_ventas"].fillna(data["salon_ventas"].mean())
data["tarjetas_debito"] = data["tarjetas_debito"].fillna(data["tarjetas_debito"].mean())
```

```python
#Sustituir valores nulos con la MEDIANA
##Mediana (median): Se utiliza cuando los datos tienen outliers o una distribución sesgada, ya que es menos sensible a valores extremos.
data["tarjetas_credito"] = data["tarjetas_credito"].fillna(data["tarjetas_credito"].median())
data["otros"] = data["otros"].fillna(data["otros"].median())
```

```python
#Sustituir valores nulos con un VALOR ESPECÍFICO (or ejemplo, 0)
data["otros_medios"] = data["otros_medios"].fillna(1028013.285)
```

```python
#Rellenar valores nulos con el VALOR ANTERIOR (forward fill)
data["subtotal_ventas_alimentos_bebidas"] = data["subtotal_ventas_alimentos_bebidas"].fillna(method="ffill")
data["bebidas"] = data["bebidas"].fillna(method="ffill")
data["almacen"] = data["almacen"].fillna(method="ffill")
data["panaderia"] = data["panaderia"].fillna(method="ffill")
data["panaderia"] = data["panaderia"].fillna(method="ffill")
data["lacteos"] = data["lacteos"].fillna(method="ffill")
data["carnes"] = data["carnes"].fillna(method="ffill")
data["verduleria_fruteria"] = data["verduleria_fruteria"].fillna(method="ffill")
data["alimentos_preparados_rotiseria"] = data["alimentos_preparados_rotiseria"].fillna(method="ffill")
```

```python
##Si queremos sustituir nulos por valores numericos
##string en concreto
data["indumentaria_calzado_textiles_hogar"]=data["indumentaria_calzado_textiles_hogar"].fillna("772871.2876")
data["electronicos_articulos_hogar"]=data["electronicos_articulos_hogar"].fillna("1628423.28")
```

```python
##Identificar los valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
data.to_csv('Actividad 3.1.csv', index=False)
```
